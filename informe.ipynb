## Consigna

El objetivo del trabajo práctico es plantear un modelo de lot sizing para planificar la producción en una fábrica. La planificación es mensual, y consiste en determinar qué cantidad
fabricar de cada producto en cada mes, durante los próximos 24 meses. Se tienen cinco productos y se tiene una estimación de la demanda por producto y mes (que se especifica en
la Tabla 1). Suponemos que las cantidades demandadas de todos los productos están en las
mismas unidades. La planificación está sujeta a las siguientes restricciones:
1. Cada producto tiene una ``capacidad de producción máxima de 120`` unidades por mes.

2. Los productos se deben fabricar en ``lotes de 10`` unidades cada uno. Es decir, la producción de cada producto en cada mes debe ser un número perteneciente al conjunto
{0, 10, 20, 30, . . . , 120}.

1. El depósito tiene una ``capacidad total de 900`` unidades, entre todos los productos. Se
puede guardar stock fabricado en un mes para cubrir la demanda de los meses siguientes.

1. Al inicio de la planificación no se tiene stock inicial de ningún producto.

2. La producción de una unidad de cada producto tiene un ``costo de $370`` (el mismo costo
por unidad para todos los productos).

El objetivo de la planificación es determinar qué cantidad fabricar de cada producto en cada
mes, de modo tal de cumplir la demanda minimizando los costos de fabricación. Suponemos
que no hay costos de almacenamiento de stock. Sobre la base de esta descripción, el trabajo
práctico consiste de los siguientes puntos:

1. Plantear y resolver un modelo de programación lineal entera mixta para este problema,
con el objetivo de minimizar los costos totales de producción.

2. Supongamos que tenemos un límite de producción de 300 unidades por mes (es decir, la
producción total entre todos los productos no debe superar las 300 unidades mensuales).
¿Cómo cambia el modelo en este caso? ¿Se puede cumplir con la demanda total?

3. Supongamos que además de las 300 unidades por mes, podemos tercerizar la fabricación
de hasta 200 unidades por mes, con un costo de $ 540 por unidad. ¿Cómo cambia el
modelo en este caso? (suponemos que tenemos que cumplir toda la demanda de cada
mes)

4. Supongamos además que no se puede tercerizar la fabricación de cualquier cantidad de
cada producto, sino que si se terceriza la fabricación de un producto en un mes, se deben
solicitar al menos 20 unidades de ese producto en ese mes. ¿Cómo cambia el costo total
en este caso?

5. El límite de 900 unidades en stock en el depósito es demasiado costoso y necesitamos
reducir el espacio disponible en el depósito. ¿Cuál es la mínima capacidad del depósito
que hace que el problema sea factible? ¿Se puede buscar este valor sin hacer búsqueda
binaria manualmente?

## Resolución

#### <span style=color:lightblue>1. Plantear y resolver un modelo de programación lineal entera mixta para este problema, con el objetivo de minimizar los costos totales de producción. </span>

``Desarrollo del Modelo Teórico:``

Para abordar el problema de optimización de la producción, es fundamental comenzar con la formulación de un modelo matemático que permita representar de manera precisa la situación real de la fábrica. Este modelo se construirá mediante la definición de variables, una función objetivo y las restricciones asociadas al sistema de producción.


``Definición de Conjuntos:``

A continuación, se propone una descripción detallada de cada uno de los conjuntos que utilizaremos en el trabajo:

- Conjunto de Productos (P) : 
Este conjunto representa los diferentes tipos de productos que se producen en la fábrica. Cada elemento de este conjunto corresponde a un producto distinto. La definición formal del conjunto P es la siguiente:
$$ P = \{\text{Producto A, Producto B, Producto C, Producto D, Producto E}\} $$

- Conjunto de Meses de Producción (M): 
Este conjunto representa el horizonte temporal de producción que se está considerando en el modelo. Cada elemento de este conjunto corresponde a un mes específico dentro de este horizonte. La definición formal del conjunto M es la siguiente:
$$ M = \{1, ... , 24\} $$

- Conjunto de Meses de Almacenamiento ($M_s$): 
Este conjunto representa los meses en los que se considera el almacenamiento de los productos. Es similar al conjunto de meses de producción pero incluye un periodo adicional (mes 0), lo cual representa un periodo inicial antes de comenzar la producción. En este primer periodo no se tendrá stock inicial. La definición formal del conjunto $M_s$ es la siguiente:
$$ M = \{0, ... , 24\} $$


``Definición de Parámetros:``

Se proporciona una definición detallada para cada uno de los parámetros, junto con una descripción de su significado y cómo se utilizarán en el modelo matemático:

- Demanda (d): 
Este parámetro representa la demanda de cada producto en cada mes. Es una matriz indexada por los conjuntos 
M (meses de producción) y P (productos), donde cada elemento, $d_{m,p}$ indica la demanda del producto p en el mes m. Los valores de la demanda se obtienen del archivo "Datos_final.txt".

- Costo por Unidad (costo):
Este parámetro es una constante que representa el costo de producir una unidad de cualquier producto en la fábrica. En este caso, el costo por unidad es de $370.
$$ costo = 370 $$

- Capacidad Máxima de Producción por Mes (capProd):
Este parámetro es una constante que representa la capacidad máxima de producción de la fábrica en un mes, es decir, el número máximo de unidades que se pueden producir en un mes. En este caso, la capacidad máxima de producción por mes es de 120 unidades.
$$ capProd = 120 $$

- Capacidad Máxima del Depósito (capDep):
Este parámetro es una constante que representa la capacidad máxima del depósito o almacén donde se guardan los productos. Indica el número máximo de unidades que se pueden almacenar en el depósito en cualquier momento. En este caso, la capacidad máxima del depósito es de 900 unidades.
$$ capDep = 900 $$

``Definición de Variables:``

Definiremos las siguientes variables para representar la producción de cada producto en cada mes:

- $ X_{M,P} $  Numero de unidades a fabricar de producto p en el mes m.
- $ S_{M_s,p} $ Cantidad de stock del producto p al final del mes m
- $ W_{M,P} $ Unidades enteras que luego utilizaremos para la restricción de los lotes. 

Se limitan las mismas a ser mayores o iguales a 0. 

``Función Objetivo:``

El objetivo principal es minimizar el costo total de producción. Por lo tanto, la función objetivo será la sumatoria del costo de producción de cada unidad de producto en cada mes. Dado que el costo por unidad es de $370, la función objetivo se puede expresar como:

$$ \text{Minimizar} \sum_{m \in M} \sum_{p \in P} 370 \cdot X_{m,p} $$


``Restricciones:``

Las restricciones del modelo estarán relacionadas con las capacidades de producción, demanda, y posiblemente otras condiciones operativas o logísticas que afecten la producción. Estas pueden incluir:

**Definición de Stock (defstock):**
Esta restricción establece la relación entre el stock del mes actual, el stock del mes anterior, las unidades fabricadas y la demanda del producto p en el mes m. En donde el stock del mes actual será la suma del stock del mes anterior más la diferencia entre las unidades fabricadas y la demanda del producto p en el mes actual.

$$ \forall m \in M  \text{ , } \forall p \in P: s_{m,p} = s_{m−1,p} + x_{m,p} − d_{m,p} $$

Ya que la variable de stock debe ser mayor o igual a 0, se obliga a que se cumpla la demanda y que se respete $ s_{m−1,p} + x_{m,p} ≥ d_{m,p} $.

**Capacidad de Producción Máxima de Cada Producto por Mes (maxprod):**
Esta restricción asegura que la producción de cada producto en cualquier mes no exceda la capacidad máxima de producción especificada (capProd) que en este caso es de 120 unidades:
$$ \forall m \in M  \text{ , } \forall p \in P: x_{m,p} ≤ capProd $$

**Lotes de 10 Unidades (lotes):**
Esta restricción establece que la producción del producto p en el mes m debe ser en lotes de 10 unidades, siendo, $ w_{m,p} $ la variable que representa el número de lotes:
$$ \forall m \in M  \text{ , } \forall p \in P:  x_{m,p} = 10 \cdot w_{m,p} $$

Pero, ¿Como funciona?. 
Empecemos por identificar a la variable $ w_{m,p} $, la cual representa el número de lotes. Los mismos son enteros. Para lograr que la producción de cada producto en cualquier mes sea en lotes de 10 unidades, se iguala $ x_{m,p} $ a la variable $ w_{m,p} $ por 10. Esto hará que $ x_{m,p} $ sea un múltiplo de 10, contabilizando la cantidad de producida y $ w_{m,p} $ la cantidad de lotes. Por lo tanto, si $ w_{m,p} = 1 $, entonces $ x_{m,p} = 10 $, si $ w_{m,p} = 2 $, entonces $ x_{m,p} = 20 $, y así sucesivamente.

**Capacidad Máxima del Depósito (maxstock):**
Esta restricción asegura que el stock total de todos los productos en cualquier mes no exceda la capacidad máxima del depósito especificada (capDep) que en este caso es de 900 unidades:
$$ \forall m \in M: \sum_{p \in P} s_{m,p} ≤ capDep $$
​
**Stock Inicial en Cero (stockinicial):**
Esta restricción establece que el stock de cada producto p al inicio del horizonte de planificación (mes 0) es cero:
$$ \forall p \in P: s_{0,p} = 0 $$

Para abordar esta nueva condición, se realizan varias modificaciones en el modelo de optimización.

``Parametros Agregados``

- **limite maximo de produccion en los meses (maxProd):**
Indica que la producción total entre todos los productos no debe superar las 300 unidades mensuales.
$$ maxProd = 300 $$

``Restricciones Agregadas``

- **Limite de produccion maxima mensual:** 
La restricción impone que para todo mes, la suma de lo producido entre todos los productos en ese mes, no puede ser mayor a 300:
$$ \forall m \in M: \sum_{p \in P} x_{m,p} ≤ maxProd $$

``Resultados``

Desafortunadamente, al resolver este modelo ampliado, el software de optimización informa que no existe una solución factible para el problema bajo las restricciones dadas. Esto significa que no es posible satisfacer la demanda total y, al mismo tiempo, mantener la producción total por debajo del límite de 300 unidades por mes. 

Como resultado, el modelo se vuelve infactible bajo estas condiciones:

**Solution Status: Infeasible - No solution available**


#### <span style=color:lightblue>3. Supongamos que además de las 300 unidades por mes, podemos tercerizar la fabricación de hasta 200 unidades por mes, con un costo de $ 540 por unidad. ¿Cómo cambia el modelo en este caso? (suponemos que tenemos que cumplir toda la demanda de cada mes) </span>

En este escenario, hemos considerado la posibilidad de subcontratar parte de la producción para aumentar la capacidad de satisfacer la demanda. Se nos permite terciarizar hasta 200 unidades mensuales de productos con un costo de $540 por unidad terciarizada. 

Para abordar esta nueva condición, hemos realizado varias modificaciones en nuestro modelo de optimización.

``Parametros Agregados``

- **costo de produccion tercerizada (costoTerciarizar):**
Representa el costo por unidad terciarizada, establecido en $540.
$$ costoTerciarizar = 540 $$
- **limite maximo de produccion tercerizada (maxTerciarizar):**
Indica la capacidad máxima de terciarización, fijada en 200 unidades al mes.
$$ maxTerciarizar = 200 $$

``Variables Agregadas``

- **$Z_{M,P}$** Representa la cantidad de unidades terciarizadas del producto p en el mes m. Se limita la misma a ser mayor o igual a 0 y a pertenecer en el conjunto de numeros enteros, ya que consideramos que los productos no pueden producirse en unidades fraccionarias.


``Función Objetivo Agregada``

Hemos ajustado nuestra función objetivo para minimizar los costos totales, considerando tanto los costos de producción como los costos de terciarización. Esto se logra mediante la siguiente expresión:
$$ \text{Minimizar} \sum_{m \in M} \sum_{p \in P} 370 \cdot X_{m,p} + 540 \cdot Z_{m,p} $$

``Restricciones Modificadas``

- **Definición de Stock (defstock):**
Se agrega la variable **$Z_{M,P}$** para contabilizar la producción tercearizada y ayudar a cumplir con la demanda necesaria. La restricción se modifica de la siguiente manera:
$$ \forall m \in M \text{ , } \forall p \in P: s_{m,p} = s_{m−1,p} + x_{m,p} + z_{m,p} − d_{m,p} $$


``Restricciones Agregadas``

- **Maxima fabricación de unidades terciarizadas**
Se introduce una restricción que limita la cantidad de unidades terciarizadas a no más de 200 unidades al mes.
$$ \forall m \in M: \sum_{p \in P} z_{m,p} ≤ maxTerciarizar $$

``Resultados``

Esta nueva formulación del modelo ha resultado en una solución óptima. En consecuencia, se logra satisfacer la demanda total de productos durante todos los meses. Notamos que la tercerización se ha empleado eficazmente como estrategia para sortear las restricciones de la producción interna. El costo operativo total ascendió a $3,128,900, cifra que engloba tanto los costos de producción como los de tercerización. 


**Valor Objetivo: $3,128,900**

El aumento del costo total en el punto 3 en comparación con el punto 1 se debe principalmente a la introducción de la terciarización y a la restricción de producción mensual máxima de 300 unidades. Al querer cumplir la demanda, nos vemos obligados a hacer uso de la terciarización aunque esta sea de mayor costo. Es por esta razón que nuestro costo total aumenta un 7.72%.

Se le agrega a la restricción de tercerización que si se terceriza la fabricación de un producto en un mes, se deben solicitar al menos 20 unidades de ese producto en ese mes. Esto se logra mediante las siguientes restricciones:

Esta restricción asegura que  $ v_{m,p} $ nunca será 1 si $ z_{m,p} $ es menor a 20 (minProdTerc), por lo que nunca habrán producciones menores a 20 unidades.
$$ \forall m \in M \text{ , } \forall p \in P: z_{m,p} ≥ minProdTerc \cdot v_{m,p} $$

Sin embargo, esta restricción no garantiza que para $ z_{m,p} $ mayor a 20, $ v_{m,p} $ sea siempre 1. Para lograr esto, se agrega la siguiente restricción:

- **Maxima fabricación de unidades terciarizadas:**
En donde maxProdTerc equivale a 200 que es la capacidad máxima de terciarización. Esta restricción asegura que $ v_{m,p} $ nunca será 0 si existe alguna producción tercerarizada $ z_{m,p} $.

$$ \forall m \in M \text{ , } \forall p \in P: z_{m,p} ≤ maxProdTerc \cdot v_{m,p} $$


Combinando ambas restricciones, los valores posibles para $ v_{m,p} $ en 0 y 1 son las siguientes:

$$
\begin{cases}
 0 ≤ z_{m,p} < 20 \, & \text{si solo si:  } v_{m,p} = 0\\
20 ≤ z_{m,p} ≤ 200 & \text{si solo si:  } v_{m,p} = 1\\
\end{cases}
$$

#### <span style=color:lightblue> 5. El límite de 900 unidades en stock en el depósito es demasiado costoso y necesitamos reducir el espacio disponible en el depósito. ¿Cuál es la mínima capacidad del depósito que hace que el problema sea factible? ¿Se puede buscar este valor sin hacer búsqueda binaria manualmente? </span>

En este quinto punto, se plantea una situación en la que se busca reducir el espacio máximo en el depósito con el objetivo de determinar cuál es la capacidad mínima que permite que el problema siga siendo factible. Es importante destacar que deseamos lograr esto sin tener que realizar una búsqueda binaria manual.

Para abordar esta nueva condición, hemos realizado varias modificaciones en nuestro modelo de optimización.

``Variables Agregadas``

- capMin: Representa la capacidad mínima requerida para el depósito para que haya una solución factible. Desconocida por el momento.  

``Función Objetivo Modificada``

Originalmente, la capacidad del depósito se había establecido en 900 unidades. Sin embargo, ahora queremos reducir esta capacidad para minimizar los costos de espacio de almacenamiento.

La función objetivo se ha simplificado para buscar minimizar capMin, lo que implica hallar la menor capacidad de depósito que satisface todas las restricciones y demandas. 

$$ \text{Minimizar :    capMin} $$

``Restricciones Modificadas``

- **Capacidad Máxima del Depósito (maxstock):** 
Se modifica la restricción de capacidad máxima del depósito para que sea igual a capMin, que es la variable que queremos minimizar. En lugar de un valor fijo de 900 unidades.

$$ \forall m \in M: \sum_{p \in P} s_{m,p} \leq \text{capMin} $$


``Resultados``

Luego de ejecutar el modelo de optimización, hemos alcanzado una solución óptima. La capacidad mínima del depósito que mantiene el problema factible se establece en **705 unidades**.
