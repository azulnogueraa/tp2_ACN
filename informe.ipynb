El objetivo del trabajo práctico es plantear un modelo de lot sizing para planificar la producción en una fábrica. La planificación es mensual, y consiste en determinar qué cantidad
fabricar de cada producto en cada mes, durante los próximos 24 meses. Se tienen cinco productos y se tiene una estimación de la demanda por producto y mes (que se especifica en
la Tabla 1). Suponemos que las cantidades demandadas de todos los productos están en las
mismas unidades. La planificación está sujeta a las siguientes restricciones:
1. Cada producto tiene una ``capacidad de producción máxima de 120`` unidades por mes.

2. Los productos se deben fabricar en ``lotes de 10`` unidades cada uno. Es decir, la producción de cada producto en cada mes debe ser un número perteneciente al conjunto
{0, 10, 20, 30, . . . , 120}.

1. El depósito tiene una ``capacidad total de 900`` unidades, entre todos los productos. Se
puede guardar stock fabricado en un mes para cubrir la demanda de los meses siguientes.

1. Al inicio de la planificación no se tiene stock inicial de ningún producto.

2. La producción de una unidad de cada producto tiene un ``costo de $370`` (el mismo costo
por unidad para todos los productos).

El objetivo de la planificación es determinar qué cantidad fabricar de cada producto en cada
mes, de modo tal de cumplir la demanda minimizando los costos de fabricación. Suponemos
que no hay costos de almacenamiento de stock. Sobre la base de esta descripción, el trabajo
práctico consiste de los siguientes puntos:

1. Plantear y resolver un modelo de programación lineal entera mixta para este problema,
con el objetivo de minimizar los costos totales de producción.

2. Supongamos que tenemos un límite de producción de 300 unidades por mes (es decir, la
producción total entre todos los productos no debe superar las 300 unidades mensuales).
¿Cómo cambia el modelo en este caso? ¿Se puede cumplir con la demanda total?

3. Supongamos que además de las 300 unidades por mes, podemos tercerizar la fabricación
de hasta 200 unidades por mes, con un costo de $ 540 por unidad. ¿Cómo cambia el
modelo en este caso? (suponemos que tenemos que cumplir toda la demanda de cada
mes)

4. Supongamos además que no se puede tercerizar la fabricación de cualquier cantidad de
cada producto, sino que si se terceriza la fabricación de un producto en un mes, se deben
solicitar al menos 20 unidades de ese producto en ese mes. ¿Cómo cambia el costo total
en este caso?

5. El límite de 900 unidades en stock en el depósito es demasiado costoso y necesitamos
reducir el espacio disponible en el depósito. ¿Cuál es la mínima capacidad del depósito
que hace que el problema sea factible? ¿Se puede buscar este valor sin hacer búsqueda
binaria manualmente?

``Desarrollo del Modelo Teórico:``

Para abordar el problema de optimización de la producción, es fundamental comenzar con la formulación de un modelo matemático que permita representar de manera precisa la situación real de la fábrica. Este modelo se construirá mediante la definición de variables, una función objetivo y las restricciones asociadas al sistema de producción.


``Definición de Conjuntos:``

A continuación, se propone una descripción detallada de cada uno de los conjuntos que utilizaremos en el trabajo:

- Conjunto de Productos (P) : 
Este conjunto representa los diferentes tipos de productos que se producen en la fábrica. Cada elemento de este conjunto corresponde a un producto distinto. La definición formal del conjunto P es la siguiente:
$$ P = \{\text{Producto A, Producto B, Producto C, Producto D, Producto E}\} $$

- Conjunto de Meses de Producción (M): 
Este conjunto representa el horizonte temporal de producción que se está considerando en el modelo. Cada elemento de este conjunto corresponde a un mes específico dentro de este horizonte. La definición formal del conjunto M es la siguiente:
$$ M = \{1, ... , 24\} $$

- Conjunto de Meses de Almacenamiento ($M_s$): 
Este conjunto representa los meses en los que se considera el almacenamiento de los productos. Es similar al conjunto de meses de producción pero incluye un periodo adicional (mes 0), lo cual representa un periodo inicial antes de comenzar la producción. En este primer periodo no se tendrá stock inicial. La definición formal del conjunto $M_s$ es la siguiente:
$$ M = \{0, ... , 24\} $$


``Definición de Parámetros:``

Se proporciona una definición detallada para cada uno de los parámetros, junto con una descripción de su significado y cómo se utilizarán en el modelo matemático:

- Demanda (d): 
Este parámetro representa la demanda de cada producto en cada mes. Es una matriz indexada por los conjuntos 
M (meses de producción) y P (productos), donde cada elemento, $d_{m,p}$ indica la demanda del producto p en el mes m. Los valores de la demanda se obtienen del archivo "Datos_final.txt".

- Costo por Unidad (costo):
Este parámetro es una constante que representa el costo de producir una unidad de cualquier producto en la fábrica. En este caso, el costo por unidad es de $370.

- Capacidad Máxima de Producción por Mes (capProd):
Este parámetro es una constante que representa la capacidad máxima de producción de la fábrica en un mes, es decir, el número máximo de unidades que se pueden producir en un mes. En este caso, la capacidad máxima de producción por mes es de 120 unidades.

- Capacidad Máxima del Depósito (capDep):
Este parámetro es una constante que representa la capacidad máxima del depósito o almacén donde se guardan los productos. Indica el número máximo de unidades que se pueden almacenar en el depósito en cualquier momento. En este caso, la capacidad máxima del depósito es de 900 unidades.


``Definición de Variables:``

Definiremos las siguientes variables para representar la producción de cada producto en cada mes:

- $ X_{M,P} $  Numero de unidades a fabricar de producto p en el mes m
- $ S_{M_s,p} $ Cantidad de stock del producto p al final del mes m
- $ W_{M,P} $ Unidades enteras que luego utilizaremos para la restricción de los lotes. 


``Función Objetivo:``

El objetivo principal es minimizar el costo total de producción. Por lo tanto, la función objetivo será la sumatoria del costo de producción de cada unidad de producto en cada mes. Dado que el costo por unidad es de $370, la función objetivo se puede expresar como:

$$ \text{Minimizar} \sum_{m \in M} \sum_{p \in P} 370 \cdot X_{m,p} $$


``Restricciones:``

Las restricciones del modelo estarán relacionadas con las capacidades de producción, demanda, y posiblemente otras condiciones operativas o logísticas que afecten la producción. Estas pueden incluir:

**Definición de Stock (defstock):**
Esta restricción establece la relación entre el stock del mes actual, el stock del mes anterior, las unidades fabricadas y la demanda del producto p en el mes m. En donde el stock de el mes actual será la suma del stock del mes anterior más la diferencia entre las unidades fabricadas y la demanda del producto p en el mes actual.

$$ s_{m,p} = s_{m−1,p} + x_{m,p} − d_{m,p} $$

**Capacidad de Producción Máxima de Cada Producto por Mes (maxprod):**
Esta restricción asegura que la producción de cada producto en cualquier mes no exceda la capacidad máxima de producción especificada (capProd):
$$ x_{m,p} ≤ capProd $$
